```python
def cod_idx(x):
    if pd.isna(x):
        return 0**0.11
    elif x <= 3 :
        return 100**0.11
    elif x > 3 and x <= 25 :
        return 67**0.11
    elif x > 25 and x <= 50 :
        return 34**0.11
    else :
        return 1**0.11
def clr_idx(x):
    if pd.isna(x):
        return 0**0.11
    elif x <= 1.4 :
        return 100**0.11
    elif x > 1.3 and x <= 10 :
        return 67**0.11
    elif x > 10 and x <= 20 :
        return 34**0.11
    else :
        return 1**0.11
# def ong_idx(x):
#     if pd.isna(x):
#         return 0**0.18
#     elif x == 0 :
#         return 100**0.18
#     elif x > 0 and x <= 0.2 :
#         return 67**0.18
#     elif x > 0.2 and x <= 0.3 :
#         return 34**0.18
#     else :
#         return 1**0.18
def TN_idx(x):
    if pd.isna(x):
        return 0**0.12
    elif x <= 0.25 :
        return 100**0.12
    elif x > 0.25 and x <= 0.75 :
        return 67**0.12
    elif x > 0.75 and x <= 1.5 :
        return 34**0.12
    else :
        return 1**0.12
# def fcoli_idx(x):
#     if pd.isna(x):
#         return 0**0.09
#     elif x <= 500 :
#         return 100**0.09
#     elif x > 500 and x <= 2000 :
#         return 67**0.09
#     else :
#         return 1**0.09
def TP_idx(x):
    if pd.isna(x):
        return 0**0.12
    elif x <= 0.02 :
        return 100**0.12
    elif x > 0.02 and x <= 0.5 :
        return 67**0.12
    elif x > 0.5 and x <= 1 :
        return 34**0.12
    else :
        return 1**0.12
def TSS_idx(x):
    if pd.isna(x):
        return 0**0.18
    elif x <= 20 :
        return 100**0.18
    elif x > 20 and x <= 100 :
        return 67**0.18
    else :
        return 1**0.18
```
